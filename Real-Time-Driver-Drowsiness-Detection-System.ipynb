```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join('data','images')
labels = ['awake','drowsy']
number_imgs = 100
```

```python
cap = cv2.VideoCapture(0)

for label in labels:
    print(f"\nNOW COLLECTING: {label.upper()} ")

    for i in range(5, 0, -1):
        ret, frame = cap.read()
        cv2.putText(frame, f"Get ready: {label.upper()} ({i})",
                    (50, 100), cv2.FONT_HERSHEY_SIMPLEX,
                    1.2, (0, 0, 255), 3)
        cv2.imshow('Image Collection', frame)
        cv2.waitKey(1000)

    for img_num in range(number_imgs):
        ret, frame = cap.read()

        cv2.putText(frame, f"{label.upper()}",
                    (50, 100), cv2.FONT_HERSHEY_SIMPLEX,
                    2, (0, 255, 0), 4)

        imgname = os.path.join(IMAGES_PATH, label + '.' + str(uuid.uuid1()) + '.jpg')
        cv2.imwrite(imgname, frame)

        cv2.imshow('Image Collection', frame)

        if cv2.waitKey(500) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path = 'yolov5/runs/train/exp/weights/last.pt')
```

```python
import time
import winsound
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

drowsy_start_time = None
no_face_start_time = None
alert_triggered = False

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    detections = results.xyxy[0]

    label = "No Detection"
    current_time = time.time()

    # FACE DETECTED
    if len(detections) > 0:
        cls = int(detections[0][5])
        label = results.names[cls]

        no_face_start_time = None  # reset no-face timer

        # DROWSY
        if label == "drowsy":
            if drowsy_start_time is None:
                drowsy_start_time = current_time

            elapsed = current_time - drowsy_start_time

            if elapsed >= 3:
                alert_triggered = True

            if alert_triggered:
                winsound.Beep(1000, 300)
                time.sleep(0.3)

            cv2.putText(frame, f"Drowsy: {int(elapsed)}s",
                        (50, 50), cv2.FONT_HERSHEY_SIMPLEX,
                        1, (0, 0, 255), 2)

        else:
            drowsy_start_time = None
            alert_triggered = False

            cv2.putText(frame, "Status: Awake",
                        (50, 50), cv2.FONT_HERSHEY_SIMPLEX,
                        1, (0, 255, 0), 2)

    # NO FACE DETECTED
    else:
        drowsy_start_time = None

        if no_face_start_time is None:
            no_face_start_time = current_time

        elapsed_no_face = current_time - no_face_start_time

        if elapsed_no_face >= 3:
            winsound.Beep(800, 300)
            time.sleep(0.3)

        cv2.putText(frame, f"No Face: {int(elapsed_no_face)}s",
                    (50, 50), cv2.FONT_HERSHEY_SIMPLEX,
                    1, (0, 255, 255), 2)

    # OUTPUT
    cv2.imshow('Live Detection', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
